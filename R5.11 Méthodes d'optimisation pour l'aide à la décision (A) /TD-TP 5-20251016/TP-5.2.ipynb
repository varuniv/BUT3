```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# # Set the random seed for reproducibility
# torch.manual_seed(42)
# np.random.seed(42)
```

```python
# Définir l'architecture du réseau de neurones
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(2, 2)  # Première couche (taille d'entrée : 2, taille de sortie : 2)
        self.fc2 = nn.Linear(2, 2)  # Deuxième couche (taille d'entrée : 2, taille de sortie : 2)

        # Initialize weights manually
        self.fc1.weight.data = torch.tensor([[0.0, 1.0], [0.0, -1.0]])  # W1
        self.fc2.weight.data = torch.tensor([[1.0, 2.0], [2.0, -1.0]])  # W2

        # ATTENTION : On doit initialiser biases pour reseau a 0 
        # car par default le reseau ajouter les biases aleatoires pour
        # tous les couches cachees implicitement. on peut revenir
        # au graph de calcul pour introduire l'ajout de biases pour les couches cachees
        # Les biases pour les couches cachees sont presentees dans la CM a la fin semaine.
        self.fc1.bias.data = torch.zeros(2)  # Initialize biases for fc1
        self.fc2.bias.data = torch.zeros(2)  # Initialize biases for fc2
      

    def forward(self, x):
        u1 = self.fc1(x)          # Transformation linéaire
        u2 = torch.relu(u1)      # Activation ReLU
        u3 = self.fc2(u2)        # Transformation linéaire
        h = torch.softmax(u3, dim=0)  # Activation softmax pour obtenir les probabilités
        return h  # Retourner les probabilités
```

```python

model = SimpleNN() 
#Definir le loss. Ce loss de Pytorch equivalent au notre code Loss = -np.sum(y * np.log(h + 1e-10))
criterion = nn.BCELoss() 

# # Le loss suivant de Pytorch include implicitement la fonction softmax,
# # alors on doit supprimer la partie  h = torch.softmax(u3, dim=0) si on veut utiliser ce loss
# # Vous pouvez essayer
# criterion = nn.CrossEntropyLoss()  # Cross-entropy loss for multi-class classification

optimizer = optim.SGD(model.parameters(), lr=0.01)  # Stochastic Gradient Descent

x = torch.tensor([1.0, 1.0])  
# y = torch.tensor([0])  # (0 for class 1, 1 for class 2) si on utilise nn.CrossEntropyLoss()
y = torch.tensor([1.0, 0.0])  #

# Training loop for a single iteration
for iteration in range(1):
    # Forward pass
    output = model(x)
    # loss = criterion(output.unsqueeze(0), y)  # si on utilise nn.CrossEntropyLoss()
    loss = criterion(output.unsqueeze(0), y.unsqueeze(0))
    
    # Backward pass
    optimizer.zero_grad()  # remettre les gradients dans les previous iterations aux 0
    loss.backward()  # Backpropagation
    optimizer.step()  # Update weights

    print(f"Iteration {iteration + 1}:")
    print("Output after forward pass:", output)
    print("Cross-entropy loss before update:", loss.item())
    print("Updated weights W1:\n", model.fc1.weight.data)
    print("Updated weights W2:\n", model.fc2.weight.data)
   
```
